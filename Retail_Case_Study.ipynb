```python
#1. Merge the datasets Customers, Product Hierarchy and Transactions as Customer_Final. 
#Ensure to keep all customers who have done transactions with us and select the join type accordingly.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

Customers = pd.read_csv(r"C:\Users\gagan\Downloads\Python Foundation Case Study 1 - Retail Case Study\Case Study 1 - Retail Case Study\Customer.csv")
Product_Hierarchy = pd.read_csv(r"C:\Users\gagan\Downloads\Python Foundation Case Study 1 - Retail Case Study\Case Study 1 - Retail Case Study\prod_cat_info.csv")
Transactions = pd.read_csv(r"C:\Users\gagan\Downloads\Python Foundation Case Study 1 - Retail Case Study\Case Study 1 - Retail Case Study\Transactions.csv")

customer_transaction = pd.merge(Transactions, Customers, how="inner", left_on="cust_id", right_on="customer_Id")

customer_final = pd.merge(customer_transaction, Product_Hierarchy, how="inner", on="prod_cat_code")

customer_final.head()

```

```python
# 2. Prepare a summary report for the merged data set.
# a. Get the column names and their corresponding data types

column_summary = customer_final.dtypes
print("Summary Report: Column Names and Data Types")
print(column_summary)
```

```python
#b. Top/Bottom 10 observations

customer_final_sorted = customer_final.sort_values(by="total_amt", ascending=False)

top_10 = customer_final_sorted.head(10)
print("Top 10 Observations:")
top_10
```

```python
bottom_10 = customer_final_sorted.tail(10)
print("Bottom 10 Observations:")
bottom_10
```

```python
#c. “Five-number summary” for continuous variables (min, Q1, median, Q3 and max)

continuous_vars = customer_final.select_dtypes(include=["int64", "float64"])

five_number_summary = continuous_vars.describe().loc[["min", "25%", "50%", "75%", "max"]]
five_number_summary.rename(index={"min": "Minimum", "25%": "Q1", "50%": "Median", "75%": "Q3", "max": "Maximum"}, inplace=True)

print("Five-Number Summary:")
five_number_summary
```

```python
#d. Frequency tables for all the categorical variables

categorical_columns = ['Gender', 'city_code', 'prod_cat', 'prod_subcat', 'Store_type']

for column in categorical_columns:
    print(f"Frequency table for {column}:\n")
    print(customer_final[column].value_counts())
    print("\n")
```

```python
# 3. Generate histograms for all continuous variables and frequency bars for categorical variables.
import matplotlib.pyplot as plt
import seaborn as sns

# Define continuous and categorical columns
continuous_columns = ['Qty', 'Rate', 'Tax', 'total_amt']
categorical_columns = ['Gender', 'city_code', 'prod_cat', 'prod_subcat', 'Store_type']

# Generate histograms for continuous variables
for column in continuous_columns:
    plt.figure(figsize=(8, 5))
    sns.histplot(customer_final[column], kde=True, bins=30, color='blue')
    plt.title(f'Histogram of {column}', fontsize=14)
    plt.xlabel(column, fontsize=12)
    plt.ylabel('Frequency', fontsize=12)
    plt.grid(axis='y')
    plt.show()

# Generate frequency bar charts for categorical variables
for column in categorical_columns:
    plt.figure(figsize=(8, 5))
    sns.countplot(data=customer_final, x=column, palette='viridis')
    plt.title(f'Frequency Bar Chart of {column}', fontsize=14)
    plt.xlabel(column, fontsize=12)
    plt.ylabel('Count', fontsize=12)
    plt.xticks(rotation=45)
    plt.grid(axis='y')
    plt.show()
```

```python
# 4. Calculate the following information using the merged dataset :
# a. Time period of the available transaction data

customer_final['tran_date'] = pd.to_datetime(customer_final['tran_date'], format='%m/%d/%Y', errors='coerce')

start_date = customer_final['tran_date'].min()
end_date = customer_final['tran_date'].max()

print(f"Time period of the available transaction data: {start_date.date()} to {end_date.date()}")
```

```python
# b. Count of transactions where the total amount of transaction was negative

negative_transactions = customer_final[customer_final['total_amt'] < 0]

negative_transaction_count = negative_transactions.shape[0]

print(f"Count of transactions with negative total amount: {negative_transaction_count}")
```

```python
# 5. Analyze which product categories are more popular among females vs male customers.

gender_category_sales = customer_final.groupby(['Gender', 'prod_cat'])['total_amt'].sum().reset_index()

popular_categories = gender_category_sales.sort_values(by=['Gender', 'total_amt'], ascending=[True, False])

popular_categories

```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a bar plot to compare popularity of product categories by gender
plt.figure(figsize=(12, 6))
sns.barplot(data=gender_category_sales, x='prod_cat', y='total_amt', hue='Gender', palette='Set2')
plt.title('Product Category Popularity by Gender', fontsize=16)
plt.xlabel('Product Category', fontsize=12)
plt.ylabel('Total Sales Amount', fontsize=12)
plt.xticks(rotation=45)
plt.legend(title='Gender')
plt.tight_layout()
plt.show()
```

```python
# 6. Which City code has the maximum customers and what was the percentage of customers from that city?

city_counts = customer_final['city_code'].value_counts()

max_city_code = city_counts.idxmax()
max_city_count = city_counts.max()

total_customers = city_counts.sum()

percentage = (max_city_count / total_customers) * 100

print(f"City code with the maximum customers: {max_city_code}")
print(f"Number of customers from that city: {max_city_count}")
print(f"Percentage of customers from that city: {percentage:.2f}%")
```

```python
# 7. Which store type sells the maximum products by value and by quantity?

store_analysis = customer_final.groupby('Store_type').agg(
    total_quantity=('Qty', 'sum'),
    total_value=('total_amt', 'sum')
).reset_index()

max_quantity_store = store_analysis.loc[store_analysis['total_quantity'].idxmax()]
max_value_store = store_analysis.loc[store_analysis['total_value'].idxmax()]

print("Store type with maximum products sold by quantity:")
print(f"Store Type: {max_quantity_store['Store_type']}, Quantity: {max_quantity_store['total_quantity']}\n")

print("Store type with maximum products sold by value:")
print(f"Store Type: {max_value_store['Store_type']}, Value: {max_value_store['total_value']}\n")
```

```python
# 8. What was the total amount earned from the "Electronics" and "Clothing" categories from Flagship Stores?
filtered_data = customer_final[
    (customer_final['prod_cat'].isin(['Electronics', 'Clothing'])) &
    (customer_final['Store_type'] == 'Flagship Stores')
]

total_amount_earned = filtered_data['total_amt'].sum()

print(f"Total amount earned from 'Electronics' and 'Clothing' categories in Flagship Stores: {total_amount_earned}")
```

```python
# 9. What was the total amount earned from "Male" customers under the "Electronics" category?

filtered_data = customer_final[
    (customer_final['Gender'] == 'Male') &
    (customer_final['prod_cat'] == 'Electronics')
]

total_amount_male_electronics = filtered_data['total_amt'].sum()

print(f"Total amount earned from Male customers under the Electronics category: {total_amount_male_electronics}")

```

```python
# 10. How many customers have more than 10 unique transactions, after removing all transactions which have any negative amounts?

filtered_data = customer_final[customer_final['total_amt'] >= 0]

transaction_counts = filtered_data.groupby('cust_id')['transaction_id'].nunique()

customers_with_10_plus_transactions = transaction_counts[transaction_counts > 10]

num_customers = customers_with_10_plus_transactions.shape[0]

print(f"Number of customers with more than 10 unique transactions: {num_customers}")
```

```python
# 11. For all customers aged between 25 - 35, find out:
# a. What was the total amount spent for “Electronics” and “Books” product categories?
from datetime import datetime

current_year = datetime.now().year
customer_final['age'] = current_year - pd.to_datetime(customer_final['DOB']).dt.year

age_filtered_data = customer_final[(customer_final['age'] >= 25) & (customer_final['age'] <= 35)]

category_filtered_data = age_filtered_data[
    age_filtered_data['prod_cat'].isin(['Electronics', 'Books'])
]

total_amount_spent = category_filtered_data['total_amt'].sum()

print(f"Total amount spent on Electronics and Books by customers aged 25-35: {total_amount_spent}")
```

```python
# b. What was the total amount spent by these customers between 1st Jan, 2014 to 1st Mar, 2014
from datetime import datetime

current_year = datetime.now().year
customer_final['age'] = current_year - pd.to_datetime(customer_final['DOB'], dayfirst=True).dt.year
age_filtered_data = customer_final[(customer_final['age'] >= 25) & (customer_final['age'] <= 35)]

customer_final['tran_date'] = pd.to_datetime(customer_final['tran_date'], dayfirst=True)

date_filtered_data = age_filtered_data[
    (customer_final['tran_date'] >= '2014-01-01') &
    (customer_final['tran_date'] <= '2014-03-01')
]

total_amount_spent = date_filtered_data['total_amt'].sum()

print(f"Total amount spent by customers aged 25–35 between 1st Jan 2014 and 1st Mar 2014: {total_amount_spent}")
```
